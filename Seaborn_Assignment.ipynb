## 2: Plot 12 most dangerous cities

```python
c = kill.city.value_counts().head(12)
c
```

```python
# with seaborn barplot
fig, ax = plt.subplots(figsize=(8,6))
sns.barplot(x = c.index, y = c.values, data = kill)  
plt.xticks(rotation=45)
ax.set_title("Most Dangerous Cities", color = "blue", fontsize = 15)
```

```python
# with seaborn countplot
plt.figure(figsize=(8,5))
plt.title('Most Dangerous Cities',color ='blue',size = 15)
ax.set_xticklabels(kill['city'].value_counts().head(12).index, rotation=45);
sns.countplot(x = 'city', data =kill, order = kill['city'].value_counts().head(12).index)
plt.xticks(rotation=45);
```

## 3: Plot 20 most dangerous states

```python
s = kill.state.value_counts().head(20)
s
```

```python
fig, ax = plt.subplots(figsize=(7.7,5))
sns.barplot(x = s.index, y = s.values, data = kill)  
ax.set_xticklabels(s.index, rotation=45);
ax.set_title("Most Dangerous states", color = "blue", fontsize = 15);
```

```python
# with seaborn countplot
plt.figure(figsize=(7.7,5))
plt.title('Most Dangerous states',color ='blue',size = 15)
ax.set_xticklabels(kill['state'].value_counts().head(12).index, rotation=45);
sns.countplot(x = 'state', data =kill, order = kill['state'].value_counts().head(20).index)
plt.xticks(rotation=45);
```
