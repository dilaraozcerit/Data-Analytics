```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
labels = ['UK', 'USA', 'GER', 'JAP', 'FRA']
age_men = [40, 35, 45, 53, 28]
age_women = [35, 32, 48, 48, 31]
ind = np.arange(len(labels))
```

```python
fig, ax = plt.subplots()
bar1 = ax.bar(labels, age_men, label='Men', width=0.5)

ax.set_title("Age means by Gender")
ax.set_ylabel("Age")

ax.legend()

ax.bar_label(bar1)

fig.tight_layout()

```

```python
fig, ax = plt.subplots()
bar2 = ax.bar(labels,age_women,label='Women', width=0.5)

ax.set_title("Age means by Gender")
ax.set_ylabel("Age")

ax.legend()

ax.bar_label(bar2)

fig.tight_layout()
```

```python
thikness=0.3
x=np.arange(len(labels))
fig, ax = plt.subplots()

bar1=ax.bar(x-1/2*thikness,age_women,label='Women', width=thikness)
bar2=ax.bar(x+1/2*thikness,age_men,label='Men', width=thikness)

ax.set_title("Age means by Gender")
ax.set_ylabel("Age")

ax.set_xticks(x)
ax.set_xticklabels(labels)

ax.legend()

ax.bar_label(bar1)
ax.bar_label(bar2)

fig.tight_layout()
```
