```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,40,20,15],
                   'var2': [100,253,333,262,111,969, 232,343,123]})
df6
```

```python
df6[["var1", "var2"]].transform
```

```python
df6[["var1"]].transform(lambda x : ( x - df6["var1"].min() )/
                                       (df6["var1"].max()-df6["var1"].min())  # advance kullanım yaptık normalde bu kadar uzun değil.
                       )
```

```python
df6[["var1"]].transform(lambda x : (x - x.min()) / (x.max() - x.min()))
```

```python
# 1) transform() works with function, a string function, a list of functions, and a dict. However, apply() is only allowed with function.
# (2) transform() cannot produce aggregated results.
# (3) apply() works with multiple Series at a time. But, transform() is only allowed to work with a single Series at a time.
```

```python
df6.groupby("groups")[["var1", "var2"]].apply(np.mean)
```

```python
df6.groupby("groups")[["var1","var2"]].mean()
```

```python
df6[["var1"]].transform([np.sqrt, np.exp])
```

```python
# filtering

df6.groupby("groups").filter(lambda x: x["var2"].mean() > 200 )
```

```python
df6[ df6["var2"] > 200 ]
```
