```python
import pandas as pd
import numpy as np
import scipy
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
plt.figure(figsize=(12,8))

ax = sns.scatterplot(x="total_bill", y="tip", data=tips)

ax.set(xlabel='Total Bill', ylabel='Tip', title='Some title')
#plt.legend(labels=['legend1'])
plt.show()
```

Hue   #####Hue = Groupby

```python
# Gruplamada uc tur var, hue, size veya style kullanabiliriz.
```

```python
tips.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex", s=200)
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips,hue="sex", s=200, style="sex");
```

```python
sns.scatterplot(x='total_bill',y='tip',size="sex", data=tips); # this is not a common usage
```

```python
# Alpha, Yogunlasmanin nerede oldugunu gostermek acisindan kullaniliyor.
sns.scatterplot(x='total_bill',y='tip',s=200, data=tips, alpha=0.5);  # alpha default'u 1.0 dır
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='total_bill',y='tip', data=tips,style='sex',hue='sex',s=100)
# Call savefig in the same cell
plt.savefig('example_scatter.jpg')
```

Rugplot

```python
sns.rugplot(tips["total_bill"])
```

```python
sns.rugplot(x = "total_bill", data= tips, height= 0.5) # 0.5 demek yarısı oluyor.
```

Displot and Hisplot

```python
sns.displot(x = "total_bill", data=tips);
```

```python
sns.displot(x = "total_bill", data=tips, kde = True);
```

```python
sns.histplot(data=tips, x="total_bill", bins=30, kde= True);
```

```python
sns.set(style="darkgrid")
sns.histplot(data = tips, x = "total_bill", bins=20)
```

```python
sns.displot(data = tips, x = 'total_bill',bins=20,kde=False,
            color='red',edgecolor='black',lw=4,ls='--')
```

```python
sns.set(style='darkgrid', rc={"grid.color": ".6", "grid.linestyle": ":"})
sns.histplot(data = tips, x = 'total_bill',bins=20);
```

The Kernel Density Estimation Plot

```python
#KDE
sns.kdeplot(tips.total_bill);
```

```python
sns.kdeplot(x = "total_bill", data=tips);
```

```python
sns.kdeplot(x = "total_bill", data=tips, clip = [10,20]); # kırptık.
```

Bandwith

```python
sns.kdeplot(x = "total_bill", data=tips, bw_adjust=0.2); # bw_Adjust default'u 1dir. 1'e geldikçe smoothlaşır.
```

```python
sns.kdeplot(x = "total_bill", data=tips, bw_adjust=0.2, shade= True, color="purple"); # shade içini boyar.
```

2 Dimensional KDE Plots

```python
sns.kdeplot(data=data, x = "x", y = "y", shade=True, cbar=True);
```
