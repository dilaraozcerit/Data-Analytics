```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43, 42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]

plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather")

plt.plot(days, max_t, label = "max")  # these labels won't be seen if we don't add "plt.legend() function."
plt.plot(days, min_t, label =  "min")
plt.plot(days, avg_t, label = "avg")

plt.legend(loc = "upper right")    # loc vermezsek default olarak sol aa ke alr
```

BAR CHART

```python
company = ["GOOGL", "AMZN", "MSFT", "FB"]
revenue = [90, 136, 89, 27]
```

```python
plt.bar(company, revenue)
```

```python
ypos = np.arange(len(company))
ypos
```

```python
plt.bar(ypos, revenue)
```

```python
plt.xticks(ypos, company) # numaralar tekrardan stringe dntrdk 
plt.bar(ypos, revenue)
```

```python
profit = [40, 2, 34, 12]
```

```python
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos, revenue, label = "Revenue")
plt.bar(ypos, profit, label = "Profit")
plt.legend()
```

```python
plt.xticks(ypos, company)
plt.ylabel("revenue(bln)")
plt.title("US Tech Stocks")
plt.bar(ypos-0.2, revenue, width = 0.4, label = "Revenue")
plt.bar(ypos-0.2, profit, width = 0.4, label = "Profit")
plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title("US Tech Stocks")
plt.barh(ypos-0.2, revenue, label = "Revenue")
plt.bar(ypos+0.2, profit, label = "Profit")
plt.legend()
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins = [80, 100, 125], rwidth = 0.95, color= "g")   
```

```python
bs_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
bs_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel("sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood sugar analysis")
plt.hist([bs_men, bs_women], bins = [80, 100, 125, 150], rwidth = 0.95, color= ["green", "orange"], label = ["men", "women"])   
plt.legend()
```

PIE CHART


```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utililites"]
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct= "%0.2f%%") # 2f = noktadan sonraki gsterilen say.
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct= "%0.1f%%", shadow = True) # 2f = noktadan sonraki gsterilen say.
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct= "%0.1f%%", shadow = True, explode = [0, 0.5, 0,0,0]) 
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct= "%0.1f%%", shadow = True, explode = [0, 0.5, 0,0,0.3]) 
plt.show()
```
