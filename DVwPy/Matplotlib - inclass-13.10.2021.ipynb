**Functional Method**

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
plt.plot(age, salary)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age")
```

**Object Oriented Method**

```python
fig, ax = plt.subplots()
ax.plot(age,salary, 'r')  # object-oriented is fancier.
ax.set_xlabel("age")
ax.set_ylabel('salary')
ax.set_title("Salary by Age")
plt.show()
```



---



**Multiple-Line in the same graph**

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
plt.plot(age, salary)
plt.plot(age, salary_2)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("Salary by Age");
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
plt.plot(x,y)
plt.plot(x, z)
plt.plot(x,t)           #functional hali
plt.xlabel("x")
plt.ylabel("y,z,t")
plt.title("x ' e göre y,z,t");
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.plot(x, z)   # object-oriented hali
ax.plot(x, t)
ax.set_xlabel("X")
ax.set_ylabel("Y - Z")
ax.set_title("Title")
```

**Adding Axes**

```python
fig = plt.figure()      # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])
ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")
ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

```python
fig = plt.figure() 
ax1 = fig.add_axes([1, 0.0, 1, 1])             # left, bottom, width, height
ax2 = fig.add_axes([0, 0.3, 0.4, 0.4])
ax1.plot(age,salary, "r")                  
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")                     # Buyuk figurun en alt sol kosesi 0, 0 oluyor
ax1.set_title("Salary by Age")
ax2.plot(age,salary_2, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

**Line 

```python
x = np.arange(0, 10)
y = 2*x
y
```

```python
fig, ax = plt.subplots(figsize=(12,6))
fig.suptitle("To change the line width, we can use the `linewidth` or `lw` keyword argument",fontsize=18)

ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10)
plt.show()
```

```python
# possible linestype options ‘--‘, ‘–’, ‘-.’, ‘:’, ‘steps’
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# Use marker for string code
# Use markersize or ms for size
ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20)
plt.show()
```

**Plot Range and Adding Extra Lines**

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])
plt.show()
```

```python
x  =np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))

ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')

ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')

ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])

ax[0].axvline(x=5,ls="--")
ax[1].axhline(y=15,ls="--")
plt.show()

```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[0].axvline(x=2,ls="--")
ax[0].axhline(y=20,ls="--")
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])
ax[1].axvline(x=2,ls="--")
ax[1].axhline(y=20,ls="--")
plt.show()
```

```python
# rowspan and colspan

plt.figure()
a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan =2, colspan=2)

x = np.arange(1, 10)

a1.plot(x, np.exp(x))
a1.set_title("exp")

a2.plot(x, x**2)
a2.set_title("square")

a3.plot(x, np.log(x))
a3.set_title("log")

plt.tight_layout()

plt.show()
```
