```python
# Basic Libraries
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar");
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar", estimator=np.count_nonzero);
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar", estimator=np.sum);
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar", estimator=np.sum, hue= "sex");
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar", estimator=np.sum, hue= "sex", col="alone", row= "embarked");
```

```python
sns.catplot(x= "pclass", y= "survived", data = titanic, kind="bar", estimator=np.sum, hue= "sex", col="sex");
```

```python
tips= sns.load_dataset("tips")
tips.head(1)
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "box");
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "box", hue = "smoker");
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "violin", hue = "smoker");
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "swarm", hue = "smoker");
```

```python
sns.catplot(x = "day", y = "total_bill", data = tips, kind = "bar", estimator = np.count_nonzero, hue = "smoker");
```

**Joinplot**

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="hex")
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind="kde")
```

**Pairplot**

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = "sex");
```

```python
sns.pairplot(tips, hue = "sex", palette = "viridis", corner=True);
```

```python
sns.pairplot(tips, hue = "sex", palette = "viridis", corner=True, diag_kind= 'hist'); # distribution yerine hist getirir.
```

**PairGrid**

```python
g=sns.PairGrid(tips)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, color ="r");
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend()
```

```python
a = sns.pairplot(tips)
a.map_lower(sns.kdeplot, color='purple')
a.map_upper(sns.kdeplot, color='g');
```

**Facetgrid**

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(data = tips, col="time", row="smoker")
g = g.map(plt.scatter, "total_bill", "tip")
```
