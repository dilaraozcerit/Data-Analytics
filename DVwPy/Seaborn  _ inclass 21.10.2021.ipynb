```python
# Basic Libraries
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset ("tips")
tips.head()
```

```python
tips.describe()
```

**Countplot**

```python
tips.day.value_counts()
```

```python
sns.countplot(x= "day", data= tips);
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+1));
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+1))
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) )     # y√ºzdelik ekledik
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips, hue="sex")
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+0.7))
    ax.annotate(round(p.get_height()/tips.day.count(),2), (p.get_x()+0.1,
                                   p.get_height()-4))
```

```python
flights = sns.load_dataset("flights")
```

```python
sns.lineplot(x = "year", y="passengers", data= flights)
```

```python
f_sum = flights.groupby(["year", "month"]).sum()
f_sum
```

```python
sns.lineplot(y=f_sum.passengers, x = f_sum.reset_index().index)
```

```python
flights_wide = flights.pivot(index = "year", columns = "month", values = "passengers")
flights_wide
```

```python
sns.lineplot(data=flights_wide);
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```
