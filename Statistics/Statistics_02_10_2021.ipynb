```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
tv_hours = [3,5,2,0,5,3,1,4,3,4]
GPA = [2.7,2.1,3.3,3.4,2.0,3.0,3.6,2.8,3.5,2.6]
```

```python
tv_hours * 3
```

```python
tv_hours = np.array(tv_hours)
GPA = np.array(GPA)
```

```python
tv_hours * 3
```

```python
np.corrcoef(tv_hours, GPA)  # bir sıralaması yok, tersini de yazabilirsin
 # tv ilk sütun ikinci GPA
 # ilk row tv ikinci GPA
 
```

Correlation with Pandas

```python
carcrash = sns.load_dataset("car_crashes")
```

```python
carcrash.head()
```

```python
carcrash.describe()
```

```python
carcrash.corr() # 100000 ler bir korelasyon oluşturuyor.
```

```python
carcrash.corr().loc["total","no_previous"]
```

```python
carcrash["total"].corr(carcrash["no_previous"])
```

```python
carcrash.corr(method="spearman")
```

```python
sns.heatmap(carcrash.corr());
```

```python
sns.heatmap(carcrash.corr(), cmap="RdYlGn"); # farklı bir methodu
```

```python
sns.heatmap(carcrash.corr(), cmap="RdYlGn", annot= True);
```

```python
sns.pairplot(carcrash)
```
