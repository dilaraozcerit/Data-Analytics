**Semantic Variables**

```python
sns.set_style("darkgrid")
```

```python
# hue
sns.scatterplot(x = "carat", y = "price",
                data = diamonds);
```

```python
sns.scatterplot(x="carat", y="price", hue='cut',
                data=diamonds);
```

```python
sns.scatterplot(x="carat", y="price", hue='depth',
                data=diamonds);
```

```python
sns.scatterplot(x="carat", y="price", hue='depth',
                data=diamonds,
                s=100);
```

```python
sns.scatterplot(x="carat", y="price", size='cut',
                data=diamonds,
                s=100,
                alpha = 0.4
                );
```

**Hıstplot**

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins.dropna(inplace=True)
```

```python
sns.set_style('darkgrid')
sns.histplot(penguins.bill_length_mm);
```

```python
sns.set_style('darkgrid')
sns.histplot(y = 'bill_length_mm', data = penguins); # same as above but horizontal.
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, bins= 20);
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, binwidth= 8, binrange=(30, 60));
```

```python
# statistics part

sns.histplot(x = 'bill_length_mm', data = penguins, stat="count");
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, stat="density");
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, stat="probability");
```

```python
sns.histplot(x = 'bill_length_mm', data = penguins, stat="probability", fill=False, element='step', cumulative=True); # shows how probability accumulates as we move across this x-axis
```

```python
#categorical variables

sns.histplot(x='bill_length_mm', data= penguins, hue= 'species', element='step');
```

```python
sns.histplot(x='bill_length_mm', data= penguins, hue= 'species', multiple='stack');
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data= penguins);
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data= penguins, hue= 'species');
```

```python
#styling

sns.histplot(x='species', hue= 'sex', multiple='dodge', data=penguins, shrink=0.8); #shrink  = gap between each of bars
```

```python
sns.histplot(x='species', hue= 'sex', multiple='dodge', data=penguins, shrink=0.8, fill= False);
```

KDEplot


```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.kdeplot(cars.horsepower);
```

```python
sns.kdeplot(cars.horsepower, shade= True);
```

```python
sns.kdeplot(cars.horsepower, cumulative=True);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels=20);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, n_levels=20, shade= True);
```

```python
sns.set_style('whitegrid')
sns.kdeplot(cars.horsepower, cars.mpg, n_levels=20, shade= True, shade_lowest= False);
```

```python
sns.kdeplot(cars.horsepower, cars.mpg, shade= True, shade_lowest= False, cbar=True);
```
