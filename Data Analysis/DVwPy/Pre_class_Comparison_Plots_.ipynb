Catplot:
seaborn.catplot(x=None, y=None, hue=None, data=None, row=None, col=None, 
estimator=function, palette=None, legend=True)
*The kind parameter selects the underlying axes-level function to use: (box, boxen, point, bar, violin, swarm...)*

```python
cars = sns.load_dataset("mpg").dropna()
```

```python
cars.head()
```

```python
sns.set_style('white')
```

```python
sns.catplot(x = "horsepower", data = cars);
```

```python
sns.catplot(x = "horsepower", data = cars, kind = "box");
```

```python
sns.catplot(x = "origin", y = "horsepower",
            data = cars,
            hue = "cylinders",
            kind = "box");
```

```python
sns.catplot(x = "horsepower", data = cars, kind = "violin");
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "box",
            col = "origin");
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "strip",
            col = "origin");
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "strip",
            col = "origin",
           );
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "strip",
            col = "origin",
            row_order = ['new', 'old']  # row_order >>> we put the new cars on the top row and the old cars in the bottom row.
           );
```

## Styling

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "box",
            col = "origin",
            palette = "winter"); # palette changes colors.
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "box",
            col = "origin",
            palette = "winter",
            whis = 1);    # whis >> whisker length   ( default is 1.5 ) # now we can see more outliers.
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "strip",
            col = "origin");
```

```python
sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "strip",
            col = "origin",
            s = 12,   #  s >> increases the size of dots.
            alpha = 0.3,  # alpha >> transparency
            linewidth = 5  # linewidth >> adds more line.
            );
```

```python
sns.set(style = "darkgrid", font_scale= 1.5) # font_scale

g =sns.catplot(x = "cylinders", y = "horsepower",
            data = cars,
            kind = "box",
            col = "origin");

g.set_titles(col_template = "{col_name} cars"); # col_template
```

## JointPlot:
seaborn.jointplot(x=None, y=None, data=None, kind='scatter',  hue=None)

bivariate and univariate graphs.


**Plot kinds :**
1) scatter (default)
2) kde
3) reg
4) hist
5) hex
6) resid

```python
geyser = sns.load_dataset("geyser")
```

```python
geyser.head()
```

```python
sns.jointplot(x = geyser.waiting, y = geyser.duration);
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser);  # exact same plot
```

```python
# Plot kinds : 
#1) scatter (default) 
#2) kde 
#3) reg 
#4) hist 
#5) hex 
#6) resid
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser);  # bkz. scatter plot by default.
```

```python
 sns.jointplot(x = "waiting", y = "duration", data = geyser, kind = "kde" );    # kde >> kernel density estimation
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser, kind = "reg" );
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser, kind = "hist" );
```

```python
g = sns.jointplot( x = "waiting", y = "duration", data = geyser);
```

```python
type(g)
```

```python
g.plot_joint(sns.kdeplot, color = "gray", levels = 5);
```

```python
### categorical variables

geyser.head()
```

```python
#hue
sns.jointplot(x = "waiting", y = "duration", data = geyser, hue = "kind");
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser,
              hue = "kind",  # ( note = hue >> the column in dataframe is called kind )
              kind = "kde"   # kind >> argument of the jointplot
              );
```

## Styling

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser,
              color = "green");
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser,
              hue = "kind",
              palette = "autumn");
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser,
              height = 8,
              space = 0.2, # space >> makes marginal distribution closer to main figure
              ratio = 5); # ratio >> how big the main figure compared to the marginals. default ratio is 5.
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser,
              height = 8,
              space = 0.2,                                                                        # ratio !!
              ratio = 10);
```

```python
# joint_kws

sns.jointplot(x = "waiting", y = "duration", data = geyser,
              joint_kws = {'marker' : 8});         # marker : 5>> triangles point to the right      
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser, 
              kind = "reg");
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser, 
              kind = "reg",
              joint_kws = {'ci': None, 'order':2});
```

```python
# marginal_kws
sns.jointplot(x = "waiting", y = "duration", data = geyser, 
              marginal_kws = {"color" : "xkcd:golden"});
```

```python
sns.jointplot(x = "waiting", y = "duration", data = geyser, 
              marginal_kws = {"lw" : 4});  # we changed linewidth.
```

## Pairplot:
seaborn.pairplot(data, hue=None, palette=None, kind='scatter')

Plot pairwise relationships in a dataset.

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.pairplot(tips);
```

```python
tips.dtypes
```

```python
sns.pairplot(tips);
```

```python
sns.pairplot(tips, diag_kind='kde');
```

```python
sns.pairplot(tips, kind = 'reg');
```

```python
## categoric

tips['weekend'] = tips.day.isin(["Sat", "Sun"])
tips.head()
```

```python
# hue

sns.pairplot(tips, hue = 'weekend');
```

```python
tips.dtypes # seaborn treats boolean values as numeric.
```

```python
sns.pairplot(tips, hue = "weekend", vars = ["total_bill", "tip"]);
```

```python
sns.pairplot(tips,
             hue = "weekend",
             y_vars = ["tip"],
             x_vars = ["total_bill", "size"],
             kind = "reg",
             height = 4);
```

Styling

```python
from ipywidgets import interactive

def size_widget (height = 2.5, aspect = 1):
  sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widget, height = (1,3.5,0.5), aspect = (0.5, 2, 0.25))                      # wooooww :))
```

```python
sns.pairplot(tips, hue= "day", palette = "plasma");
```

```python
sns.pairplot(tips, diag_kws = {'color': 'gray', 'alpha' : 0.6}); # diag_kws >> passes whatever keyword arguments you put in the dictionary
```

```python
sns.pairplot(tips, kind= 'reg', plot_kws={"ci":None, 'color': "xkcd:salmon"});
```
